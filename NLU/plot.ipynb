```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
size=(8, 5)
```

```python
atis_bert = pd.read_csv("trained_models/atis_model_bert/training curves.csv")
atis_roberta = pd.read_csv("trained_models/atis_model_roberta/training curves.csv")
snips_bert = pd.read_csv("trained_models/snips_model_bert/training curves.csv")
snips_roberta = pd.read_csv("trained_models/snips_model_roberta/training curves.csv")
```

```python
atis_bert.columns = ['intent_acc', 'loss', 'semantic_frame_acc', 'slot_f1', 'slot_precision','slot_recall']
atis_roberta.columns = ['intent_acc', 'loss', 'semantic_frame_acc', 'slot_f1', 'slot_precision',
                      'slot_recall']
snips_bert.columns = ['intent_acc', 'loss', 'semantic_frame_acc', 'slot_f1', 'slot_precision',
                      'slot_recall']
snips_roberta.columns = ['intent_acc', 'loss', 'semantic_frame_acc', 'slot_f1', 'slot_precision',
                      'slot_recall']
```

```python
atis_bert.columns
```

```python
metrics_plotted = ['intent_acc', 'slot_f1', 'semantic_frame_acc']
```

```python
fig, g = plt.subplots(1, 1, figsize=size)
# plt.setp(g, xticks=ks, xticklabels=ks)
g.set_title("Learning curve on Atis")
g.set_ylabel('Loss')
g.set_xlabel('Epochs')
g.plot(atis_bert['loss'],
       label='loss'+"-BERT", marker='o')
g.plot(atis_roberta['loss'],
       label='loss'+"-RoBERTa", marker='o')
g.legend(shadow=True, fancybox=True)
```

```python
fig, g = plt.subplots(1, 1, figsize=size)
# plt.setp(g, xticks=ks, xticklabels=ks)
g.set_title("Learning curve on Snips")
g.set_ylabel('Loss')
g.set_xlabel('Epochs')
g.plot(snips_bert['loss'],
       label='loss'+"-BERT", marker='o')
g.plot(snips_roberta['loss'],
       label='loss'+"-RoBERTa", marker='o')
g.legend(shadow=True, fancybox=True)
```

```python
fig, g = plt.subplots(1, 1, figsize=size)
# plt.setp(g, xticks=ks, xticklabels=ks)
g.set_title("Metrics on Atis")
g.set_ylabel('Loss')
g.set_xlabel('Epochs')
for idx in range(len(metrics_plotted)):
    g.plot(atis_bert[metrics_plotted[idx]],
           label=metrics_plotted[idx]+"-bert", marker='o')
    g.legend(shadow=True, fancybox=True)
for idx in range(len(metrics_plotted)):
    g.plot(atis_roberta[metrics_plotted[idx]],
           label=metrics_plotted[idx]+"-roberta", marker='o')
    g.legend(shadow=True, fancybox=True)
```

```python
fig, g = plt.subplots(1, 1, figsize=size)
# plt.setp(g, xticks=ks, xticklabels=ks)
g.set_title("Metrics on Snips")
g.set_ylabel('Loss')
g.set_xlabel('Epochs')
for idx in range(len(metrics_plotted)):
    g.plot(snips_bert[metrics_plotted[idx]],
           label=metrics_plotted[idx]+"-bert", marker='o')
    g.legend(shadow=True, fancybox=True)
for idx in range(len(metrics_plotted)):
    g.plot(snips_roberta[metrics_plotted[idx]],
           label=metrics_plotted[idx]+"-roberta", marker='o')
    g.legend(shadow=True, fancybox=True)
```
